```
import numpy as np
'''
    The pattern and weight vector will be represented as matrix of array.
    Numpy is a powerful library which supports the use of array. Hence,
    numpy.array will be used to convert list into array.
'''

def count_network(pattern, weight_vector):
  '''
    This function counts the network value by summing the multiplication 
    of pattern and weight vector.
    Numpy has a built in function - numpy.dot, in which calculate the
    dot product of two arrays.
  '''
  return np.dot(pattern, weight_vector)

def activation_function(network):
  '''
    This function determines the output of the calculated network value.
    If network value is larger than 0, then return 1
    else return 0
  '''
  if (network > 0):
    return 1
  else:
    return 0

def update_weights(pattern, target, output, weight_vector, learning_rate):
  '''
    This function updates the weights in each learning process.
  '''
  return ((target - output) * learning_rate * pattern)

n_training_set = int(input("Input jumlah training_set: "))
bias = '1' # Set bias to be 1. String data type is chosen since it will be
           # appended to a list and later on will be converted into array with
           # integer data type.

training_set = [] # Create a current list for training set.
targets = [] # Create a current list for target values.

for i in range(n_training_set):
  print("====== Data {} ======".format(i+1))
  pat = input("Input 3 perceptron value (not including bias): ") + bias
  tar = input("Input Target: ")
  training_set.append(list(pat))
  targets.append(tar)
  print()

training_set = np.array(training_set, dtype=int) # Convert the training_set 
                                                 # into numpy array.
targets = np.array(targets, dtype=int) # Convert the targets into numpy array.
weights = np.zeros(4, dtype=int) + 1 # Set the initial weights to be zero for all 

learning_rate = 1 # Set the learning rate to be 1.

epochs = 1 # Initialize the epoch count.
recursive = True # Initialize a counter to True.
while(recursive):
  print("=====Epochs {}=====".format(epochs))
  output_list = []

  for i in range(0, len(training_set[0])):
    print("Pattern: ", training_set[i])
    print("Target: ", targets[i])
    print("Weight Vector", weights)
    
    network = count_network(training_set[i], weights)
    print("Network: ", network)

    output = activation_function(network)
    print("Output: ", output)
    output_list.append(output)

    delta_weights = update_weights(training_set[i], targets[i], output, weights, learning_rate)
    print("Delta W: ", delta_weights)

    weights += delta_weights
    
    print()

  output_array = np.array(output_list, dtype=int)
  if (np.array_equal(targets, output_array)):
    recursive = False
  epochs += 1

print()
print("Total epochs needed to converge is {}".format(epochs-1))
```
